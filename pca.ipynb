```python
x1_x2 = []
x1_x3 = []
x2_x3 = []


for i in range(len(x1)):
    element1 = np.multiply(x1[i],x2[i])
    x1_x2.append(element1)
    element2 = np.multiply(x1[i],x3[i])
    x1_x3.append(element2)
    element3 = np.multiply(x2[i],x3[i])
    x2_x3.append(element3)

x1_x2

cov12 = np.mean(x1_x2)-mean_x1*mean_x2
cov13 = np.mean(x1_x3)-mean_x1*mean_x3
cov23 = np.mean(x2_x3)-mean_x2*mean_x3

var_x1 = np.var(x1)
var_x2 = np.var(x2)
var_x3 = np.var(x3)

```

```python
x__1 = [var_x1,cov12,cov13]
x__2 = [cov12,var_x2,cov23]
x__3 = [cov13,cov23,var_x3]

x_dataframe = pd.DataFrame([x__1,x__2,x__3])
x_dataframe
```
